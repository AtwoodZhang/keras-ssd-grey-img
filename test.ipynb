```python
import numpy as np

p1 = 7326
p2 = 8514
basic = 10320
fee = 50
ratio_1 = 0.03
ratio_2 = 0.10
x1 = 5000
x2 = 8000

scale1 = 8514.0/15840.0
scale2 = 1-scale1

def over8(ba):
    return (ba-x2)*ratio_2+3000*ratio_1

def over5(ba):
    return (ba-x1)*ratio_1

possible = 0
for part_1 in range(basic):
    amount = 0
    if part_1>x2:
        amount_1 = over8(part_1)
        part_2 = basic - part_1
        if part_2>x2:
            amount_2 = over8(part_2)
        elif part_2>x1:
            amount_2 = over5(part_2)
        else:
            amount_2 = 0
    elif part_1>x1:
        amount_1 = over5(part_1)
        part_2 = basic - part_1
        if part_2>x2:
            amount_2 = over8(part_2)
        elif part_2>x1:
            amount_2 = over5(part_2)
        else:
            amount_2 = 0   
    elif part_1>=0:
        amount_1 = 0
        part_2 = basic - part_1
        if part_2>x2:
            amount_2 = over8(part_2)
        elif part_2>x1:
            amount_2 = over5(part_2)
        else:
            amount_2 = 0   
    amount = amount_1 + amount_2
    # print("now: ", "part_1: ", part_1, amount_1, "// part_2: ", part_2, amount_2, "  all: ", amount_1 + amount_2)
    if abs(amount-fee)<10:
        if abs(part_1/basic - scale1) <=0.08 or amount_1 == p1 or amount_1 == p2:
            possible += 1
            print(possible, "now: ", "part_1: ", part_1, amount_1, "// part_2: ", part_2, amount_2, "  all: ", amount, part_1/basic, scale1)
        # print(possible, "now: ", "part_1: ", part_1, amount_1, "// part_2: ", part_2, amount_2, "  all: ", amount_1 + amount_2, "scale:",part_1/basic, scale1)
print("finish")
         
```
