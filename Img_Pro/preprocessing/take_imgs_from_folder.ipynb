```python
import os
import random
import shutil

def random_select_images(source_folder, destination_folder, num_images):
    """从源文件夹中随机选择特定数量的图片并移动到目标文件夹"""
    
    # 获取源文件夹中所有的图片文件
    image_files = [f for f in os.listdir(source_folder) if f.endswith(('png', 'jpg', 'jpeg'))]
    total_images = len(image_files)

    # 检查是否有足够的图片可供选择
    if num_images > total_images:
        print(f"源文件夹中只有 {total_images} 张图片，无法选取 {num_images} 张。")
        return
    
    # 随机选择指定数量的图片
    selected_images = random.sample(image_files, num_images)

    # 如果目标文件夹不存在，创建它
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # 将选中的图片移动到目标文件夹
    for image in selected_images:
        src_path = os.path.join(source_folder, image)
        dst_path = os.path.join(destination_folder, image)
        shutil.move(src_path, dst_path)
        print(f"已移动图片: {src_path} 到 {dst_path}")

# 指定源文件夹和目标文件夹路径
source_folder = r"/home/zhangyouan/桌面/zya/dataset/681/PCScreen_Book_PhoneScreen/train/PhoneScreen/"  # 源文件夹路径
destination_folder = r"/home/zhangyouan/桌面/zya/dataset/681/PCScreen_Book_PhoneScreen/val/PhoneScreen/"  # 目标文件夹路径
num_images = 200  # 需要随机选取的图片数量

# 执行随机选择和复制操作
random_select_images(source_folder, destination_folder, num_images)
```
