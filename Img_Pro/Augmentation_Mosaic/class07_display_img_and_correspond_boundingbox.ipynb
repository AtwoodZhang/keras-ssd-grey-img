```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import json

# 文件夹路径
folder_path = 'resized_labeled_images'

# 获取所有的jpg文件
jpg_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg')]

# 随机选择一个jpg文件
random_file = random.choice(jpg_files)

# 获取图像路径和对应的json文件路径
image_path = os.path.join(folder_path, random_file)
json_path = os.path.join(folder_path, random_file.replace('.jpg', '.json'))

# 读取图像
image = plt.imread(image_path)

# 读取标签
with open(json_path, 'r') as json_file:
    labels = json.load(json_file)

# 创建图像和坐标轴
fig, ax = plt.subplots(1)

# 显示图像
ax.imshow(image)

# 绘制bounding box和标签
for label in labels:
    left = label['left']
    top = label['top']
    width = label['width']
    height = label['height']

    rect = patches.Rectangle((left, top), width, height, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    # 显示标签
    label_text = label['label']
    ax.text(left, top, f'{label_text}', color='r', fontsize=18, verticalalignment='top')

# 显示图像和bounding box
plt.show()
```
