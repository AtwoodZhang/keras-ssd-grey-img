```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# 读取图像
img_path = './240719_103320_00000slice.png'  # 替换为你的图像路径
image = Image.open(img_path)

# 定位框的位置与置信度
bbox = [84, 29, 52, 39, 94]  # 替换为你的位置数据
x1, y1, x2, y2, conf = bbox

# 创建绘图
fig, ax = plt.subplots(1)

# 显示图像
ax.imshow(image)

# 计算矩形框的位置和尺寸
rect_x = x1
rect_y = y1
rect_width = x2 - x1
rect_height = y2 - y1

# 创建矩形框
rect = patches.Rectangle((rect_x, rect_y), rect_width, rect_height, linewidth=2, edgecolor='r', facecolor='none')

# 添加矩形框到图像
ax.add_patch(rect)

# 在框上方显示置信度
plt.text(x1, y1 - 1, f'Conf: {conf}', color='red', fontsize=12)

# 显示结果
plt.show()
```
