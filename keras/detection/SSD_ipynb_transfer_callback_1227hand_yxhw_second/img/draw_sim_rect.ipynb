```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# 
img_path = './240719_103320_00000slice.png'  # 
image = Image.open(img_path)

# 
bbox = [84, 29, 52, 39, 94]  # 
x1, y1, x2, y2, conf = bbox

# 
fig, ax = plt.subplots(1)

# 
ax.imshow(image)

# 
rect_x = x1
rect_y = y1
rect_width = x2 - x1
rect_height = y2 - y1

# 
rect = patches.Rectangle((rect_x, rect_y), rect_width, rect_height, linewidth=2, edgecolor='r', facecolor='none')

# 
ax.add_patch(rect)

# 
plt.text(x1, y1 - 1, f'Conf: {conf}', color='red', fontsize=12)

# 
plt.show()
```
